```python
import os
import pandas as pd
import re
```

```python
customer_reviews = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/archive_Twitter/twcs/twcs.csv')
```

```python
customer_reviews.head()
```

```python
customer_reviews_text = customer_reviews[['text']]
```

```python
customer_reviews_text.head()
```

```python
customer_reviews_text['text_lower'] = customer_reviews_text['text'].str.lower()
```

```python
customer_reviews_text.head()
```

```python
def remove_digits(text):
    result = re.sub(r'\d', '', text)
    return result
```

```python
customer_reviews_text['text_no_digits'] = customer_reviews_text['text_lower'].apply(remove_digits)
```

```python
customer_reviews_text.head()
```

```python
import nltk

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
customer_reviews_text['text_no_stopwords'] = customer_reviews_text['text_no_digits'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
customer_reviews_text.head()
```

```python
def remove_special_characters(text):
    result = re.sub(r'[^a-zA-z0-9]', ' ', text)
    return result
```

```python
customer_reviews_text['cleaner_text'] = customer_reviews_text['text_no_stopwords'].apply(remove_special_characters)
```
